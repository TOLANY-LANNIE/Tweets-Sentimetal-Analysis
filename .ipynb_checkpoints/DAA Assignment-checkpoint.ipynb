# Punctuation removal

```python
def cleanTweets(tweet):
    tweet = re.sub(r'[^\w\s]','',tweet)
    tweet = re.sub(r'@[A-Za-z0-9]+','',tweet)
    tweet = re.sub(r'[ð]','',tweet)
    tweet = re.sub(r'[à]','',tweet)
    tweet = re.sub(r'[â]','',tweet)
    # remove all single characters
    tweet = re.sub(r'\s+[a-zA-Z]\s+', ' ',tweet)
    # Remove single characters from the start
    tweet = re.sub(r'\^[a-zA-Z]\s+', ' ',tweet)
    # Substituting multiple spaces with single space
    tweet= re.sub(r'\s+', ' ', tweet, flags=re.I)
   
     # Converting to Lowercase
    tweet = tweet.lower()
    return tweet
```

```python
#apply cleaning function to test set
test_set['clean tweet']= test_set['tweet'].apply(cleanTweets)
test_set.tail()
```

```python
#apply cleaning function to train set
data_set['clean tweet']= data_set['tweet'].apply(cleanTweets)
data_set.head()
```

# Remove stop words

```python
stop_words = stopwords.words('english')
stop_words.append('user')
stop_words.append('ur')
stop_words.append('dont')
stop_words.append('cant')
stop_words.append('isnt')
stop_words.append('sjw')
stop_words.append('amp')
stop_words.append('sta')
stop_words.append('i_am')
stop_words
```

```python
def remove_sw(tweet):
    tweet_clean = [word for word in tweet if word not in stop_words]
    return tweet_clean
```

```python
data_set['no sw tweet'] = data_set['tokenize tweet'].apply(remove_sw)
data_set[['label','tokenize tweet','no sw tweet']][0:10]
```

```python
test_set['no sw tweet'] = test_set['tokenized tweet'].apply(remove_sw)
test_set[['tokenized tweet','no sw tweet']][0:10]
```

# Stemming

```python
def stemming(tweet):
    text =[ps.stem(word) for word in tweet]
    return (" ".join(text)).strip()
```

```python
data_set['stemmed tweet'] = data_set['no sw tweet'].apply(stemming)
data_set[['label','no sw tweet','stemmed tweet']][0:10]
```

```python
test_set['stemmed tweet'] = test_set['no sw tweet'].apply(stemming)
test_set[['no sw tweet','stemmed tweet']][0:10]
```

# Lemmatisation

```python
def lemmatizar(tweet):
    lemmatized_words = [lzr.lemmatize(word) for word in tweet]
    return lemmatized_words
```

```python
data_set['lemmetized tweet'] = data_set['no sw tweet'].apply(lemmatizar)
data_set[['label','no sw tweet','lemmetized tweet']][0:10]
```

# Most common hashtags used in tweets

```python
# function to collect hashtags
def hashtag_extract(x):
    hashtags = []
    # Loop over the words in the tweet
    for i in x:
        ht = re.findall(r"#(\w+)", i)
        hashtags.append(ht)

    return hashtags
```

```python
# extracting hashtags from positive tweets
HT_regular = hashtag_extract(data_set['tweet'][data_set['label'] == 0])

# extracting hashtags from negative tweets
HT_negative = hashtag_extract(data_set['tweet'][data_set['label']  == 1])


# unnesting list
HT_regular = sum(HT_regular,[])
HT_negative = sum(HT_negative,[])
```

# 1. The frequency of positive

```python
a = nltk.FreqDist(HT_regular)
d = pd.DataFrame({'Hashtag': list(a.keys()),
                  'Count': list(a.values())})

# selecting top 20 most frequent hashtags     
d = d.nlargest(columns="Count", n = 20) 
plt.figure(figsize=(16,5))
ax = sns.barplot(data=d, x= "Hashtag", y = "Count")
ax.set(ylabel = 'Count')
plt.show()
```

# 2. The frequency of negative

```python
a = nltk.FreqDist(HT_negative)
d = pd.DataFrame({'Hashtag': list(a.keys()),
                  'Count': list(a.values())})

# selecting top 20 most frequent hashtags     
d = d.nlargest(columns="Count", n = 20) 
plt.figure(figsize=(16,5))
ax = sns.barplot(data=d, x= "Hashtag", y = "Count")
ax.set(ylabel = 'Count')
plt.show()
```
